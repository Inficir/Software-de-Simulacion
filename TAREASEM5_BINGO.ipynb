```python
# Bingo ejercicio | UN JUGADOR
import random

def crear_lista_numeros(n):
    return list(range(1, n + 1))

def crear_tabla_bingo(m, numeros_disponibles):
    return random.sample(numeros_disponibles, m)

def extraer_bola(numeros_disponibles):
    if numeros_disponibles:
        numero_extraido = random.choice(numeros_disponibles)
        numeros_disponibles.remove(numero_extraido)
        return numero_extraido
    else:
        return None

def hay_ganador(tabla):
    return all(number == 'x' for number in tabla)

def mostrar_tabla(nombre, tabla):
    print("Bingo Card de", nombre + ":", tabla)

def obtener_nombre_y_edad(jugador_numero):
    while True:
        nombre = input(f"Jugador {jugador_numero}, por favor ingresa tu Nombre: ")
        edad_str = input("Ingresa tu Edad: ")
        if edad_str.isdigit():
            edad = int(edad_str)
            if edad < 16:
                print("No puedes jugar!, debes tener al menos 16 años para poder jugar.")
                return None, None  # Regresar valores None para nombre y edad
            else:
                return nombre, edad
        else:
            print("Error!. Ingresa una Edad (numérica) válida.")

def jugar_solo():
    #se crea los numeros aleatorios de 0 hasta el 50
    n = 50
    bingo_numtot = crear_lista_numeros(n)
    m = 5

    nombre_jugador, edad_jugador = obtener_nombre_y_edad(1)

    if nombre_jugador is None or edad_jugador is None:
        return  # Regresar al menú si la edad no es válida.

    tabla = crear_tabla_bingo(m, bingo_numtot)
    
    print("\n")
    mostrar_tabla(nombre_jugador, tabla)



    while True:
        pregunta = input("¿Quieres sacar una bola? (s/n): ").strip().lower()

        if pregunta == "s":
            numero_extraido = extraer_bola(bingo_numtot)
            if numero_extraido is not None:
                
                print("Número extraído:", numero_extraido)

                if numero_extraido in tabla:
                    index = tabla.index(numero_extraido)
                    tabla[index] = 'x'
                    mostrar_tabla(nombre_jugador, tabla)
                    if hay_ganador(tabla):
                        print("¡Bingo! ¡Ganaste!")
                        break
                else:
                    print("¡Mala suerte! El número no pertenece a tu Bingo Card.")
                    mostrar_tabla(nombre_jugador, tabla)
            #else:
                #print("Ya se han extraído todos los números posibles. ¡El juego ha terminado!")
                #break
        elif pregunta == "n":
            break
        else:
            print("Por favor, ingresa 's' o 'n'.")

#'''

def jugar_dos_jugadores():
    n = 50
    m = 5
    intentos = 0
    max_intentos = 50
    ganador = None

    nombre_jugador1, edad_jugador1 = obtener_nombre_y_edad(1)
    nombre_jugador2, edad_jugador2 = obtener_nombre_y_edad(2)

    if (nombre_jugador1 is None or edad_jugador1 is None) or (nombre_jugador2 is None or edad_jugador2 is None):
        return  # Regresar al menú si la edad no es válida para uno de los jugadores.

    tabla_jugador1 = crear_tabla_bingo(m, crear_lista_numeros(n))
    tabla_jugador2 = crear_tabla_bingo(m, crear_lista_numeros(n))

    mostrar_tabla(nombre_jugador1, tabla_jugador1)
    mostrar_tabla(nombre_jugador2, tabla_jugador2)

    while True:
        if ganador is not None:
            break

        pregunta = input(f"¿{nombre_jugador1}, quieres sacar una bola? (s/n): ").strip().lower()

        if pregunta == "s":
            numero_extraido = extraer_bola(crear_lista_numeros(n))
            if numero_extraido is not None:
                
                print("Número extraído:", numero_extraido)

                if numero_extraido in tabla_jugador1:
                    index = tabla_jugador1.index(numero_extraido)
                    tabla_jugador1[index] = 'x'
                    mostrar_tabla(nombre_jugador1, tabla_jugador1)
                    if hay_ganador(tabla_jugador1):
                        ganador = nombre_jugador1
                        break
                else:
                    print("¡Mala suerte! El número no pertenece a tu Bingo Card.")
                    mostrar_tabla(nombre_jugador1, tabla_jugador1)
            
        elif pregunta == "n":
            break
        else:
            print("Por favor, ingresa 's' o 'n'.")

        if ganador is not None:
            break

        pregunta = input(f"¿{nombre_jugador2}, quieres sacar una bola? (s/n): ").strip().lower()

        if pregunta == "s":
            numero_extraido = extraer_bola(crear_lista_numeros(n))
            if numero_extraido is not None:
                
                print("Número extraído:", numero_extraido)

                if numero_extraido in tabla_jugador2:
                    index = tabla_jugador2.index(numero_extraido)
                    tabla_jugador2[index] = 'x'
                    mostrar_tabla(nombre_jugador2, tabla_jugador2)
                    if hay_ganador(tabla_jugador2):
                        ganador = nombre_jugador2
                        break
                else:
                    print("¡Mala suerte! El número no pertenece a tu Bingo Card.")
                    mostrar_tabla(nombre_jugador2, tabla_jugador2)
            
        elif pregunta == "n":
            break
        else:
            print("Por favor, ingresa 's' o 'n'.")

    if ganador is not None:
        print(f"¡{ganador} gana!")

def menu():
    print("El casino de la UDLA presenta: BINGO CARD!")
    print("-"*50)
    print("1. Jugar Solo\n2. Dos Jugadores\n3. Salir")


while True:
    menu()
    
    op = input("Elige una opción: ")

    if op == "1":
        jugar_solo()
    elif op == "2":
        jugar_dos_jugadores()
    elif op == "3":
        break
    else:
        print("Opción no válida. Por favor, elige 1, 2 o 3.")

    jugar_nuevamente = input("¿Quieres jugar de nuevo? (s/n): ").strip().lower()
    if jugar_nuevamente != "s":
        break
print("¡Gracias por jugar! Recoge tu recompensa con el profe David!")
```
